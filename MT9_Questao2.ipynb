```python
page = requests.get("https://bitcointalk.org/index.php?board=67.0")

soup = BeautifulSoup(page.content, 'html.parser')
tudo = soup.findAll('td', class_='windowbg3')
tudo2 = soup.findAll('td', class_='windowbg')
```

```python
contador = 1
for tdo in tudo:
    print(tdo.text)
    contador = contador + 1
    
contador2 = 1
for tdo2 in tudo2:
    print(tdo2.text)
    contador2 = contador2 + 1
```
